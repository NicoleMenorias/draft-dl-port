# **Phase 1:Narrative Report**

**November 10 - 16**
* Submit project proposal (title, background, and objectives).

<div style="text-align: justify">
This week marked an important stage in shaping the direction of our study. As a group, we focused on narrowing our research topic and building a strong foundation for the next phases. Our discussions centered on the unique challenges and opportunities that come with working on unstructured data—particularly medical audio. We recognized early on that sound-based datasets are still underexplored in healthcare research compared to image data, which is far more commonly used. This gave us a strong motivation to look into lung sound analysis, where deep learning remains promising but not yet fully developed.

Before finalizing the topic, we used a reverse approach. Instead of selecting a problem and then searching for the data, we first checked whether a high-quality dataset was available. We understood that the reliability of our study would depend heavily on the quality of the recordings. After reviewing the dataset together and confirming that it contained clear labels, sufficient samples, and meaningful sound variations, we decided that it was suitable for experimentation. This step helped us avoid future problems and gave us confidence that our study would be grounded on solid data.

We then began sharing individual ideas. Each member contributed different perspectives on how the dataset could be used. Some focused on improving the features of the audio, such as exploring raw waveforms versus spectrograms. Others proposed comparing different deep learning models or performing systematic testing of hyperparameters. These discussions allowed us to see the different possible directions for the study and to combine our ideas in a way that supports a strong experimental design.

A challenge that we faced this week was choosing the right models to include in our study. Since our time and resources are limited, we need models that are feasible to train but still meaningful for research. We also want to make sure the models are aligned with our objectives, especially regarding raw audio, which is less studied than spectrogram-based inputs. This made our decision-making more careful and deliberate, as the choice of models will influence the entire workflow—from preprocessing to evaluation.

To ensure a smooth flow in our written outputs, we scheduled a meeting dedicated to refining both the background of the study and the research objectives. During this meeting, we discussed how the background should naturally lead toward the gaps we want to address. After several revisions, we were able to create a more cohesive narrative that connects the motivation, the research gap, and the final objectives. This step helped us produce a clearer structure that will serve as the foundation for the rest of the study.

Overall, Phase 1 allowed us to develop a clearer understanding of our dataset, align our goals as a group, and prepare the groundwork for model selection. Our discussions and collaborative revisions strengthened the connection between the background, the rationale, and the objectives. In the next phase, we will proceed with finalizing the models for experimentation and begin planning the technical workflow for preprocessing, training, and evaluation.
