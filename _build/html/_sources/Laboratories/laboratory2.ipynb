# **Laboratory Task 2**

**Instruction:** Perform a single forward pass and compute the total error.

---

## **Given**

- **Input vector:**

  $$
  x =
  \begin{bmatrix}
  1\\[4pt]
  0\\[4pt]
  1
  \end{bmatrix}
  $$

- **Target:**

  $$
  y = 1
  $$

- **Hidden activation:**  
  $f(z) = \max(0, z)$  (ReLU)

- **Hidden weights (3 inputs → 2 hidden units):**

  $$
  W_{\text{hidden}} =
  \begin{bmatrix}
  0.2 & -0.3\\[4pt]
  0.4 & 0.1\\[4pt]
  -0.5 & 0.2
  \end{bmatrix}
  $$

- **Output weights:**

  $$
  W_{\text{out}} =
  \begin{bmatrix}
  -0.3\\[4pt]
  -0.2
  \end{bmatrix}
  $$

- **Biases:**

  $$
  \theta =
  \begin{bmatrix}
  -0.4\\[4pt]
  0.2\\[4pt]
  0.1
  \end{bmatrix}
  \quad
  (\theta_1, \theta_2 \text{ hidden};\ \theta_3 \text{ output})
  $$

We use squared error with $\tfrac{1}{2}$:

$$
E = \tfrac{1}{2}(y - \hat{y})^2
$$

---

### **Step 1 — Hidden Layer Pre-activations**

We have two hidden neurons. Compute each manually:

#### For hidden neuron 1:
$$
z_1 = (1)(0.2) + (0)(0.4) + (1)(-0.5) + (-0.4)
$$

$$
z_1 = 0.2 + 0 + (-0.5) - 0.4 = -0.7
$$

#### For hidden neuron 2:
$$
z_2 = (1)(-0.3) + (0)(0.1) + (1)(0.2) + (0.2)
$$

$$
z_2 = -0.3 + 0 + 0.2 + 0.2 = 0.1
$$

So,

$$
z =
\begin{bmatrix}
-0.7\\[4pt]
0.1
\end{bmatrix}
$$

---

### **Step 2 — Hidden Activations (ReLU)**

$$
h = \max(0, z)
\Rightarrow
h_1 = \max(0, -0.7) = 0,\quad
h_2 = \max(0, 0.1) = 0.1
$$

$$
h =
\begin{bmatrix}
0\\[4pt]
0.1
\end{bmatrix}
$$

---

### **Step 3 — Output Neuron**

$$
z_{\text{out}} = (-0.3)(0) + (-0.2)(0.1) + (0.1)
$$

$$
z_{\text{out}} = 0 + (-0.02) + 0.1 = 0.08
$$

Predicted output:

$$
\hat{y} = 0.08
$$

---

### **Step 4 — Compute Error**

$$
E = \tfrac{1}{2}(y - \hat{y})^2
$$

$$
E = \tfrac{1}{2}(1 - 0.08)^2 = \tfrac{1}{2}(0.92)^2 = 0.4232
$$
